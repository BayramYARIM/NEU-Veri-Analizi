# Metodoloji

Veri kümemizin etiketlerinden yola çıkarak bu problemin bir regresyon problemi olduğu tanısına vardım. Regresyon problemi, bir bağımlı değişkenin (hedef değişken) bir veya daha fazla bağımsız değişkenle ilişkisini analiz etmeyi amaçlar. Bu ilişkiyi kullanarak, bağımlı değişkenin gelecekteki değerlerini tahmin etmek için bir regresyon modeli oluşturulur.

* **Algoritmalar**

    - Linear regresion (Doğrusal regresyon): Doğrusal regresyon, bağımsız değişkenler arasındaki doğrusal ilişkiyi modellemek için kullanılır. En küçük kareler yöntemiyle, hedef değişkenin tahminini yapmak için bir doğru denklemi oluşturur. Basit doğrusal regresyon (tek bir bağımsız değişken) ve çoklu doğrusal regresyon (birden fazla bağımsız değişken) olmak üzere iki türü vardır.
    
    - Destek Vektör Regresyonu (SVR): SVR, doğrusal veya doğrusal olmayan regresyon problemlerini çözmek için kullanılan bir yöntemdir. Özellikle aykırı değerlere karşı dirençli olmasıyla bilinir. Destek vektör makineleri prensibiyle çalışır ve bir marjinal regresyon hedefler.
    
    - Random Forests (Rastgele Ormanlar): Rastgele ormanlar, birçok karar ağacını bir araya getirerek tahmin yapmak için kullanılan bir ensemble yöntemidir. Her ağaç bağımsız olarak çalışır ve sonuçları birleştirilir. Regresyon problemlerinde de etkilidir.

# Sonuçlar

Verilen R2 Skor ve Ortalama Kare Hatası (Mean Squared Error) değerlerine dayanarak, en iyi modelin Random Forests - GridSearchCV olduğunu söyleyebiliriz. Bu model, diğer modellere kıyasla daha yüksek bir R2 skoru (0.269) ve daha düşük bir Ortalama Kare Hatası değeri (1,762,386,391.19) göstermektedir.

R2 Skoru, regresyon modelinin gözlemlenen veriye ne kadar iyi uyduğunu gösterir. Yani, daha yüksek bir R2 Skoru, modelin veriyi daha iyi açıkladığını gösterir. Random Forests - GridSearchCV modelinin diğer modellere kıyasla daha yüksek bir R2 skoru göstermesi, daha iyi bir uyum sağladığını ve daha iyi tahminler yapabileceğini düşündürür.

Ortalama Kare Hatası ise, modelin gerçek değerlerle tahmin edilen değerler arasındaki farkları ne kadar iyi tahmin ettiğini gösterir. Daha düşük bir Ortalama Kare Hatası değeri, modelin daha az hata yaptığını ve daha iyi tahminler yaptığını gösterir. Random Forests - GridSearchCV modelinin diğer modellere kıyasla daha düşük bir Ortalama Kare Hatası değeri göstermesi, daha iyi bir performansa sahip olduğunu düşündürür.

Bu nedenlerle, verilen metrikler göz önüne alındığında, Random Forests - GridSearchCV modelinin en iyi model olduğunu söyleyebiliriz. Ancak, model seçimi veri setine ve problem alanına bağlı olarak değişebilir, bu yüzden farklı veri setleri ve problem alanları için performansı karşılaştırmak önemlidir.

**Verileri modele hazır hale getirme**

```python
# "Date of Transfer" isimli kolonumuzu modelin anlayacağı şekilde çeviriyoruz.
# dt_name, tarih bölümü türündeki sütunun adıdır

def add_datepart(df, dt_name, drop=True, time=False):
    "Creates new columns from our datetime column"
    
    
    dt_column = df[dt_name]
    column_dtype = dt_column.dtype
    

    targ_name = re.sub('[Dd]ate$', '', dt_name)
    
    # attributes are normally in lower case but we wrote this way because we will use it in columns' name too
    attr = ['Year', 'Month', 'Day']
    
    if time: 
        attr = attr + ['Hour', 'Minute', 'Second']
        
    
    #Sorry curse of dimensionality, maybe another time
    for a in attr: 
        df[targ_name + a] = getattr(dt_column.dt, a.lower())
        
    # how much time passed, we will divide by 10^9 because it is in the nanosecond format
    df[targ_name + 'Elapsed'] = dt_column.astype(int) // 10 ** 9
    
    if drop: 
        df.drop(dt_name, axis=1, inplace=True)
```

```python
# Sütun veri tipini datetime64 [ns] tipine çeviriyoruz.
data['Date of Transfer']=data['Date of Transfer'].astype('datetime64[ns]')
add_datepart(data, 'Date of Transfer', drop=True, time=False)
```

**Özellik seçimi**

```python
# I divide the data into "test" and "train".
tempX = data[['Property Type', 'Old/New', 'Duration', 'Town/City', 'District', 'County', 'Date of TransferYear', 'Date of TransferMonth', 'Date of TransferDay', 'Date of TransferElapsed']]
tempY = data[['Price']]

x_train, x_test, y_train, y_test  = train_test_split(tempX, tempY, train_size=0.9)
```

```python
x_train.head()
```

**Manuel null veriler ekleme**

```python
# Kendim random olarak null veriler ekliyorum

import random

for i in x_train:
    if i == 'Duration' or i == 'County' or i == 'Old/New': 
        for j in range(0,100):
            x_train[i][random.randint(0,1689)] = np.nan
```

```python
pd.isna(x_train).sum() # null kontrolu
```

```python
# Null verileri median göre dolduruyorum

for i in x_train:
    if i == 'Duration' or i == 'County' or i == 'Old/New': 
        x_train[i].fillna(x_train[i].median(), inplace=True)
```

```python
pd.isna(x_train).sum() # null veri kontrolü
```

**Makine öğrenmesi algoritmaları**

* **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
# Linear regression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# Predict
pred_lr = lr.predict(x_test)
pred_lr
```

* **Destek Vektör Regresyonu (SVR)**

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
svr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
svr.fit(x_train, y_train)
```

```python
pred_svr = svr.predict(x_test)
pred_svr
```

* **Random Forests (Rastgele Ormanlar)**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
rf = RandomForestRegressor(max_depth=2, random_state=0)
rf.fit(x_train, y_train)
```

```python
pred_rf = rf.predict(x_test)
pred_rf
```

**hiper parametre ayarlaması (GridSearchCv)**

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
lr_GridSearchCV = GridSearchCV(LinearRegression(),parameters, cv=None)
lr_GridSearchCV.fit(x_train, y_train)

pred_lr_GridSearchCV = lr_GridSearchCV.predict(x_test)
pred_lr_GridSearchCV
```

```python
parameters = {
    'n_estimators': [100, 150, 200, 250, 300],
    'max_depth': [1,2,3,4],
}
regr = RandomForestRegressor(random_state=0)

rf_GridSearchCV = GridSearchCV(regr, parameters)
rf_GridSearchCV.fit(x_train, y_train)

pred_rf_GridSearchCV = rf_GridSearchCV.predict(x_test)
pred_rf_GridSearchCV
```

**Model Değerlendirme**

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
print("R2 Score")

print("Linear regresion: ",r2_score(y_test, pred_lr))
print("Destek Vektör Regresyonu (SVR): ",r2_score(y_test, pred_svr))
print("Random Forests: ",r2_score(y_test, pred_rf))

print("Linear regresion - GridSearchCV: ",r2_score(y_test, pred_lr_GridSearchCV))
print("Random Forests - GridSearchCV: ",r2_score(y_test, pred_rf_GridSearchCV))
```

```python
print("Mean Squared Error")

print("Linear regresion: ",mean_squared_error(y_test, pred_lr))
print("Destek Vektör Regresyonu (SVR): ",mean_squared_error(y_test, pred_svr))
print("Random Forests: ",mean_squared_error(y_test, pred_rf))

print("Linear regresion - GridSearchCV: ",mean_squared_error(y_test, pred_lr_GridSearchCV))
print("Random Forests - GridSearchCV: ",mean_squared_error(y_test, pred_rf_GridSearchCV))
```
