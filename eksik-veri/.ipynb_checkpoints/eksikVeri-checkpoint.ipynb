# Kayıp (Eksik) Veri
İstatistikte, eksik veriler veya eksik değerler, bir gözlemdeki değişken için hiçbir veri değeri saklanmadığında ortaya çıkar. Eksik veriler yaygın bir durumdur ve verilerden çıkarılabilecek sonuçlar üzerinde önemli bir etkiye sahip olabilir.

# Kayıp (Eksik) Veri Mekanizmaları ( Türleri)

### Örnek


Little ve Rubin’e göre 3 farklı eksik veri yapısı vardır. Bunlar aşağıdaki
yapılardan oluşmaktadır (16)
1) Tamamen rasgele olarak kayıp (MissingCompletely at Random, MCAR)
2) Rasgele olarak kayıp (Missing at Random, MAR )
3) Rasgele olmayan kayıp (Missing Not at Random, MNAR

# Tamamen Rasgele Olarak Kayıp (MCAR)

Tamamen rasgele eksik veri (Missing Completely at Random-MCAR) mekanizmasında verilerin kayıp olma durumu gözlenen ve ya diğer eksik verilerden bağımsız olarak gerçekleşmiştir.
MCAR varsayımının karşılanması durumunda kayıp veri mekanizması ihmal edilebilirdir (16).
Bir anket çalışmasında kişilere giden ankette tesadüf olarak bazı soruların eksik olması ya da cevaplayan kişilerin soruyu görmemesi MCAR modeline örnek olabilir.


# Rasgele Olarak Kayıp

Rasgele olarak kayıp veri (Missing at Random - MAR) yapısı MCAR’a göre daha zayıf bir varsayımdır. Bir değişkendeki kayıp verilerin olasılığının, diğer bütün değişkenler kontrol altına alınması koşuluyla, bu değişkenin kendi değeri ile ilişkisiz olduğunu ifade etmektedir (16).
Bir anket çalışmasında bireylerin soruların ya da cevapların içeriğinden bağımsız olarak, başka herhangi bir sebeple(acelesi olduğundan, okumaktan sıkıldığından vb.) soruları atlaması MAR yapısına örnek olabilir.

Verilerin kayıp olma olasılığı eşit ve bağımsız.

# Rasgele Olmayan Kayıp
Diğer bir deyişle göz ardı edilemez kayıp olarak adlandırılan rasgele olmayan kayıp (Missing Not at Random - MNAR)) yönteminde bir gözlemin eksikliği doğrudan toplanan veya talep edilen veriler ile ilgilidir. Veri kümesi içinde diğer değişkenlerden tahmin edilemez (18).
Anketteki bir sorunun yanlış sorulmasından dolayı doğru cevabın çözülemediği durumlardaki kayıp durumu MNAR varsayımına örnektir(14).

2) Atama(Doldurma-Imputation) Yöntemleri

2.1 Ortalama Atama Yöntemi :

Eksik verilere değer atama yöntemlerinin en eskisi olan bu yöntem, koşulsuz ortalama değer atama olarak da bilinmektedir. Bu yöntem, bir değişkendeki eksik gözlemlere o değişkenin gözlemlenmiş değerlerinin aritmetik ortalamasının atanması şeklinde ifade edilebilir. Tam bir veri seti elde etmek için uygulama olarak oldukça pratik olmakla birlikte, bu yöntemin bazı önemli problemleri bulunmaktadır. Eksik verilere var olan serinin ortalamasının atanması, serinin değerlerinin dağılımını merkezileştirecek ve değişkenliği azaltacaktır. Bu durum varyans ve standart sapmayı etkileyecektir. Sonuçta eksik verilerin bu yöntemle tamamlandığı veri seti üzerinden yapılan varyans – kovaryans tahminleri sapmalı olmakta, dolayısıyla parametre tahminleri etkilenmektedir. Bu dezavantajı, yöntemin çok tercih edilen bir yöntem olmasına mani olmaktadır

2.2 En Son Gözlemi İleri Taşıma Yöntemi : Eksik veriye kendisinden önce gözlemlenen son değer atanır. Özellikle sağlık çalışmalarında daha çok kullanılmakta olan bu yöntemde, eksik verinin olduğu zaman dönemi için, belirli bir sürede gözlemin değerinde bir değişiklik olmadığı varsayılmaktadır. Ancak bu durum çoğu zaman geçerli değildir. Dolayısıyla bu yöntemin kullanılması, parametre tahminlerinde sapmalara neden olmaktadır. Verideki eksik gözlemlerin sayısı arttıkça, yöntemin başarı olasılığı düşecektir. Ayrıca yapılan bazı çalışmalar, bu yöntemin kullanılmasının hipotez testleri için özellikle 1. Tip hata oranını arttırdığını ortaya koymuştur

2.3 Hot Deck ve Cold Deck Yöntemi :

Nispeten ortalama atama yöntemine benzer olan bu yöntem, kayıp verilere sahip olan gözlemlere, bu gözlemelerin benzerlerindeki değerlerin atanması şeklinde ifade edilmektedir. Burada benzerlik ifadesi, gözlemlerin var olan değerleri üzerinden açıklanmaktadır. Yöntemin uygulanırken, kayıp veriye sahip gözlem için benzer gözlemleri içeren bir havuz oluşturulur, buna bağışçı havuzu (donor pool) denir. Kayıp veriye sahip gözlem, yani alıcı için, bağışçı havuzundan bir değer seçilerek atanır. Bu seçimin yapılmasının çeşitli yöntemleri bulunmaktadır. Bunlardan bir tanesi havuzdan rassal olarak seçilen bir gözlemin değerinin kayıp gözleme atanmasıdır. Burada dikkat edilmesi gereken bir nokta, bağışçı havuzunun yeterince büyük olmasıdır. Eğer havuz küçükse, yani az sayıda benzer gözlem varsa, bu durumda yöntemin uygulanması aynı değerlerin çok kez kullanılmasına neden olarak çeşitliliği azaltacaktır. Bunun sonucunda ortalama atama yöntemine benzer problemler ortaya çıkacaktır. Hot Deck yöntemi özellikle anket çalışmalarında sık kullanılan bir yöntemdir. Anket verisi üzerinden örnek verilecek olursa, örneğin gelirin eksik olduğu bir katılımcının gelir değişkenine, katılımcı ile aynı veya benzer demografik özelliklere sahip katılımcılar içinden rassal olarak seçilen bir katılımcının gelir değerinin atanması bir Hot Deck atamasıdır.


Hot Deck yöntemine benzer şekilde işleyen bir sürece sahip diğer bir yöntem de Cold Deck yöntemidir. Bu yöntemin Hot Deck yönteminden farkı, eksik veriye atama yapılacak olan benzer gözlemlerin dış kaynaktan seçilmesidir. Bir başka deyişle, eksik veriler aynı veri setinden benzer gözlemler yardımıyla değil, farklı bir veri setinden elde edilen benzer gözlemler yardımıyla doldurulur. Burada farklı bir veri seti, daha önce yapılmış benzer çalışmalardan elde edilebilir. Bu yöntem de Hot Deck yöntemine benzer dezavantajlara sahiptir ve çok sık kullanılan bir yöntem değildir.

2.4 En yakın komşu yöntemi (knn) :

Bu yöntem Hot Deck yönteminin özel bir halidir ve kimi kaynaklarda Hot Deck yöntemi altında anlatılmaktadır. Yöntemin işleyişi Hot Deck yönteminin mantığı ile aynıdır. Farklılaşan nokta ise, kayıp verilere atanacak değerler için, kayıp veriye sahip gözleme en yakın gözlemin seçilmesidir. Bu en yakın komşu olarak adlandırılır. En yakın komşu gözlemin belirlenmesi için, gözlemler arasındaki uzaklığı ölçecek bir metrik tanımlanır ve bu metrik yardımıyla en yakın komşu gözlem seçilir. Uzaklık fonksiyonunun veya algoritmasının kullanılmasıyla, eksik veriye sahip gözleme eşit mesafede birden fazla gözlem bulunması durumunda, en yakın komşuların ortalama gözlem değeri kayıp veriye atanır. Bu yöntemin ortalama atama ve rassal hot deck atamasına göre daha etkin olduğu belirtilmektedir. Ayrıca ek bilgi kullanılmasına rağmen, model tabanlı bir yöntem olmadığından, modelden kaynaklanabilecek sapmalara maruz kalınmayacaktır. Bu yöntemin MCAR mekanizması altında doğru tahminler üretebileceği ifade edilmiştir. Bununla birlikte, en yakın komşunun belirlenme şekline bağlı olarak, bazı alıcıların çok fazla yakın komşusu olabilmektedir veya bazı bağışçılar pek çok alıcıya yakın olabilmektedir. Bu durumda aynı değerlerin birçok kez kullanılması problemi ortaya çıkacaktır

2.5 Regresyon Ataması Yöntemi:

Regresyon ataması, bir değişkenin bir veya daha fazla değişkenin doğrusal kombinasyonu olarak ifade edilmesine dayanmaktadır. Eksik verilere atamak için, eksik veriye sahip değişken bağımlı değişken olarak seçilir. Bu değişkenle ilişkili olan, tam gözlemlenmiş diğer değişken veya değişkenler ise bağımsız değişkenler olarak alınır ve regresyon modeli kurulur. Ardından, tüm değişkenlerin tam gözlemlenen değerleri yardımıyla regresyon modelinin katsayıları elde edilir. Daha sonra, elde edilen regresyon denkleminde, bağımsız değişken olarak alınan değişken veya değişkenlerin ilgili değerleri yerine konarak, bağımlı değişkendeki kayıp gözlemler elde edilir. Bu süreç, eksik veriye sahip her değişken için ayrı ayrı işletilir ve her seferinde bağımsız değişkenler farklı olabilir. Burada elde edilen değerler, bağımlı değişkenin bağımsız değişkenlere koşullu ortalama değerleri olduğundan, bu yöntem aynı zamanda Koşullu Ortalama Atama olarak da isimlendirilmektedir. Eksik verilerin, ilişkili değişkenler yardımıyla bulunması söz konusu olduğundan bu yöntem elverişli bir yöntemdir. Bu yöntem eksik verili değişkenlerin varyans – kovaryans yapısını koruması bakımından avantajlıdır. MCAR mekanizması altında bu yöntemle sapmasız tahminciler elde etmek mümkündür. 

Bu avantajlarının yanı sıra yöntemin bazı dezavantajları da mevcuttur.
Öncelikle eksik verilerin değişkenler arasındaki ilişkiyi temel alarak atanması, değişkenler arasında zaten var olan ilişkileri güçlendirir. Bu durum özellikle de açıklayıcı değişkenler arasında olduğunda çoklu doğrusal bağlılık gibi bazı problemlere neden olmaktadır. Bunun yanı sıra, doğrusal regresyon modeli kullanıldığından, kayıp verilere atanan değerlerin tamamının aynı doğru üzerine düştüğü varsayılmaktadır ki bu kısıtlayıcı bir varsayımdır. Ayrıca, kayıp veriye sahip değişken için özel olarak tanımlanmış bir kural söz konusu ise, örneğin bu değişkenin değerlerinin belirli bir aralıkta olduğu koşulu var ise, regresyon sonucu elde edilen değerler bu koşulu ihlal edebilmektedir

Regresyon yönteminin dezavantajlarını azaltmak için kullanılan ek bir yöntem de Stokastik Regresyon Ataması yöntemidir. Bu yöntem, regresyon ataması yönteminde işleyen sürecin aynısına sahiptir ancak son adımda ek bir düzeltme adımı uygulanmaktadır. Eksik veriler için regresyon yardımıyla elde edilen değerlere rassal bir hata terimi eklenir. Bu terim sıfır ortalamaya ve regresyondan elde edilen kalıntılarla aynı varyansa sahip, normal dağılımlı rassal bir terimdir. Böylece aynı doğru üzerine düşen eksik veri değerleri rassal olarak değiştirilerek çeşitlilik arttırılmış olur. Stokastik regresyon ataması yöntemi MAR mekanizması altında da sapmasız tahminciler elde edilebilmektedir.

2.6 Beklenti Maksimizasyonu yöntemi : (EM-Expectation-maximization), eksik veri içeren olasılıklı yöntemlerde parametre tahmini için tasarlanmış model tabanlı bir atama tekniğidir. Beklenti maksimizasyonu, beklenti adımı (E) ve maksimizasyon (M) adımı olarak iki adımın art arda tekrarlanmasıyla gerçekleşir. E-adımı modelde eksik verinin tamamlanması üzerine parametrelerin o anki tahminlerini kullanarak bir log-olabilirlik beklentisi fonksiyonu oluşturur. M adımı parametre değerlerini log-olabilirlik beklentisini maksimize edecek şekilde günceller. Yani bu iki adımın her biri diğerinin girdisini hesaplayarak birbirini besler. 2.7 Çoklu atama yöntemi :
   
   İlk aşama Atama (Imputation) aşamasıdır. Bu aşamada öncelikle orijinal veri setinin kopyaları oluşturulur. Bu kopyaların her birinde eksik veriler farklı değerler atama yapılarak doldurulur. Bu adımda eksik verilere atama yapmak için kullanılabilecek farklı algoritmalar var olmakla birlikte, normal dağılıma sahip veriler için sıklıkla kullanılan bir yöntem Veri Çoğaltma yöntemidir. Bu yöntem Stokastik Regresyon yönteminin iteratif olarak yapılan bir benzeridir. Atama aşaması sonucunda m adet farklı tam veri seti elde edilmiş olur. Burada m sayısı arttıkça daha başarılı sonuçlar alınacağı düşünülse de, belli sayıdan sonra atama yapılan değerler benzer olmaya başlayacaktır. Bunun için 2 ile 10 aralığında bir sayı olması ideal olmaktadır. Tam veri setleri elde edildikten sonra, nispeten daha kolay olan Analiz (Analysis) aşamasına geçilir. Bu aşamada her bir tam veri setine tam veri yöntemleri uygulanarak, parametre tahminleri elde edilir. Dolayısıyla m sayıda veri setinden, her bir parametre için m adet tahmin elde edilmiş olur. Bir sonraki aşama olan Havuzlama (Pooling) aşamasında, bir önceki aşamada elde edilmiş olan parametreler ve standart sapmalar birleştirilir ve her bir parametre ile standart sapma için tek bir değer elde edilir. Bu birleştirme işlemi, örneğin model katsayıları için her bir katsayının farklı veri setlerinden gelen tahminlerinin aritmetik ortalaması alınarak yapılabilir.
